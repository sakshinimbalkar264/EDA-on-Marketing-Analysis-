```python
# Importing all the required libraries.

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Reading the dataset.

df = pd.read_csv('../Downloads/marketing_data.csv')
```

```python
# Analyzing duplicate values

df.duplicated().all()
```

```python
# Checking the NA values in each columns.

df.isnull().sum()
```

```python
# Clean up column names that contain whitespace

df.rename(columns={' Income ':'Income'}, inplace = True)
df['Income']=df['Income'].str.replace("$","",regex = False)
```

```python
# Transform Income column to a numerical

df["Income"]=df["Income"].str.replace(",","", regex = False)
df["Income"]=df["Income"].astype(float)
```

```python
# Plotting boxplot for checking the outliers.

plt.grid()  
plt.title('Income',size=15)
sns.boxplot(x='Income', data=df)
```

```python
sns.histplot(df['Income'],kde=False)
```

```python
#Impute null values in Income, using median value (to avoid skewing of the mean due to outliers):

df['Income']=df['Income'].fillna(df['Income'].median())
```

```python
# To convert string Date time into Python Date time object

df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'])
```

```python
df['Customer_Age'] = df['Dt_Customer'].dt.year - df['Year_Birth']
```

```python
df['Customer_Age']
```

```python
df['Total_Spent']=(df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds'])
df['Total_Purchases']=(df['NumDealsPurchases']+df['NumWebPurchases']+df['NumCatalogPurchases']+df['NumStorePurchases']+df['NumWebVisitsMonth'])
```

```python
df['Total_Spent']
```

```python
df['Total_Purchases']
```

```python
df['Kids_Teen_at_home'] = df['Kidhome'] + df['Teenhome']
```

```python
df['Kids_Teen_at_home']
```

```python
# Reducing unnecessary columns

df = df.drop(['ID','Kidhome','Teenhome'], axis = 1)
```
